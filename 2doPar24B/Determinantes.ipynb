```python
# importar numpy para los determinantes
import numpy as np
```

# A = (9)

```python
A = np.matrix([[9]])

# forma matricial
print (A.shape)

# determinante
print (np.linalg.det (A))
```

```python
#determinantes en Python
B = np.matrix ([[4,   -1],
               [-2,   0]])

# forma matricial
print (B.shape)

print(np.linalg.det (B))
```

```python
C = np.matrix ([[5,   0,   2],
               [3,   1,   1],
               [0,   1,   2]])

# forma matricial
print (C.shape)

np.linalg.det (C)
```

## RESOLVER:



```python
A = np.matrix([
    [3, 2, -1],
    [2, -2, 4],
    [-1, 0.5, -1]
])
b = np.matrix([[1], [-2], [0]])
```

```python
x = np.linalg.solve(A, b)
print("Solución x:", x)
```

```python
det_A = np.linalg.det(A)
print("Determinante de A:", det_A)
```
