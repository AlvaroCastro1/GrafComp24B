# Visualización de Vectores

En este cuaderno vamos a crear un codigo con Python para graficar y manipular vectores.
En cada una de las celdas presentan los códigos con sus respectivas descripciones y comentarios.


### 1. Importación de librerías necesarias
Se importan las librerías `numpy` y `matplotlib` que se utilizarán para la manipulación de vectores y su visualización gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. Ejemplo para graficar un vector

```python
# Graficamos un vector que parte del punto (1, 2) con dirección (3, 4)
plt.quiver(1, 2, 3, 4, scale_units='xy', angles="xy", scale=1)

# Establecemos los límites del gráfico para asegurar que se visualice correctamente
plt.xlim(0, 4)  # Limita el eje X de 0 a 4
plt.ylim(0, 8)  # Limita el eje Y de 0 a 8

# Dibujamos guías verticales y horizontales que cruzan el punto inicial y el punto final del vector
plt.axvline(x=1)        # Línea vertical que pasa por el origen del vector en x = 1
plt.axhline(y=2)        # Línea horizontal que pasa por el origen del vector en y = 2
plt.axvline(x=1 + 3)    # Línea vertical que pasa por la posición final del vector en x = 1 + 3
plt.axhline(y=2 + 4)    # Línea horizontal que pasa por la posición final del vector en y = 2 + 4
```

 #### 3. Graficar múltiples vectores desde el origen

```python
# Las listas [1,2] y [4,2] definen las componentes de los vectores en los ejes x e y, respectivamente
plt.quiver([0,0], [0,0], [1,2], [4,2], scale_units='xy', angles='xy', scale=1)
# para los lim de los ejes X e Y para visualizar correctamente ambos vectores
plt.xlim(0, 4)  # Limita el eje X de 0 a 4
plt.ylim(0, 4)  # Limita el eje Y de 0 a 4
```

#### 4. Graficar vectores con guías auxiliares

```python
# graficamos dos vectores donde el primero parte del punto (1,2) con direccion (1,3)
# y el segundo parte del punto (0,0) a (2,2)
plt.quiver([1, 0], [2, 0], [1, 2], [3, 2], scale_units='xy', angles='xy', scale=1)

# se establece los lim de los ejes X e Y para asegurar que los vectores se visualicen completamente
plt.xlim(0, 4)  # Limita el eje X de 0 a 4
plt.ylim(0, 6)  # Limita el eje Y de 0 a 6

# Aqui se agregan guias auxiliares en las posiciones iniciales y finales de los vectores
# Las guias se representan con lineas grises, punteadas y con opacidad

# para el punto inicial del primer vector
plt.axvline(x=1, color='gray', linestyle="--", alpha=0.3)  # vertical en x = 1
plt.axhline(y=2, color='gray', linestyle="--", alpha=0.3)  # horizontal en y = 2

# para el punto final del primer vector (x = 2, y = 5)
plt.axvline(x=2, color='gray', linestyle="--", alpha=0.3)  # vertical en x = 2
plt.axhline(y=5, color='gray', linestyle="--", alpha=0.3)  # horizontal en y = 5

# para el punto final del segundo vector (x = 2, y = 2)
plt.axvline(x=2, color='gray', linestyle="--", alpha=0.3)  # vertical en x = 2
plt.axhline(y=2, color='gray', linestyle="--", alpha=0.3)  # horizontal en y = 2


```

```python
# se hacer dos nuevos cetores con numpy
v1 = np.array ([1,2])
v2 = np.array([3,4])

# se grafican y establecen los lim
plt.quiver([0,0], [0,0], [v1[0],v2[0]],[v1[1],v2[1]],angles='xy', scale_units='xy', scale=1)
plt.xlim(0, max(v1[0], v2[0]))
plt.ylim(0, max(v1[1], v2[1]))

# con sus lineas de referencia
plt.axvline(x=1, color='gray', linestyle="--", alpha=0.3)
plt.axhline(y=2, color='gray', linestyle="--", alpha=0.3)

plt.axvline(x=3, color='gray', linestyle="--", alpha=0.3)
plt.axhline(y=4, color='gray', linestyle="--", alpha=0.3)
```

#### 5. Funcion para graficar multiples vectores

```python
def grafvecs(vecs, cols):
    # se inicia una figura de tamaño 5x5 para asegurar que las graf tengan un formato cuadrado
    plt.figure(figsize=(5, 5))
    
    # donde:
    #    vecs: Lista de vectores
    #    cols: Lista de colores correspondientes para cada vector
    
    for i in range(len(vecs)):
        vec = vecs[i]  # obtenemos cada vector de la lista vecs
        # los vectores se dibujan como flechas con sus colores respectivos
        plt.quiver(0, 0, vec[0], vec[1], angles='xy', color=cols[i], scale_units='xy', scale=1)
```

```python
# creamos 3 vectores
v1 = np.array ([1,2])
v2 = np.array([3,4])
v3 = np.array([8,4])

# con sus colores
grafvecs([v1, v2, v3], ['blue', 'red', "black"])

# y se establecen los lim
plt.ylim( max(v1[1], v2[1]) )
plt.xlim(0, max(v1[0], v2[0], v3[0]))
plt.ylim(0, max(v1[1], v2[1], v3[1]))
```
