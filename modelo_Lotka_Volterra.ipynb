# Modelo Lotka Volterra
Este modelo describe la interacción entre dos especies (presas y depredadores).
Se utiliza para **modelar** cómo las **poblaciones** de ambas especies 
### variacion # presas = aumento natural - destruccion por depredadores
```
x = ax-bxy
```
### variacion # depredadores = - muerte en ausencia de presas + aumento causado por alimentacion
```
y = -cy+exy
```
### Parámetros del modelo
- Tasa de crecimiento de la población presas (aumento natural)
    a = 0.6
- Tasa a la que la población del depredador reduce la población de las presas (destruccion de los depredadores)
    b = 0.4
- Tasa de disminución de la población del depredador (muerte en ausencia de presas)
    c = 0.6
- Tasa a la que la población y del depredador aumenta debido a la depredación exitosa dela presa (aumento causado por alimentacion disponible)
    e = 0.4
- Incremento de tiempo en cada paso de la simulación.
    dt = 0.001
- Tiempo máximo de la simulación
    max_time = 10
- tiempo inicial y poblaciones
    t = 0; x = 1.5; y = 1.5


````python
def lotka_volterra(a, b, c, e, dt, t0, max_time, x0, y0):
    t = t0
    x = x0
    y = y0

    # Listas para almacenar el tiempo y las poblaciones
    t_list = []
    x_list = []
    y_list = []

    # Añadir los valores iniciales a las listas
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

    # Bucle de simulación
    while t < max_time:
        # Calcular nuevos valores para t, x, y
        t = t + dt
        x = x + (a*x - b*x*y)*dt
        y = y + (-c*y + e*x*y)*dt

        # Almacenar los nuevos valores en las listas
        t_list.append(t)
        x_list.append(x)
        y_list.append(y)

    # Graficar los resultados
    plt.plot(t_list, x_list, 'r', label='Presa (x)')
    plt.plot(t_list, y_list, 'g', label='Depredador (y)')
    plt.legend()
    plt.grid(True)
    plt.xlabel('Tiempo')
    plt.ylabel('Población')
    plt.title('Modelo Lotka-Volterra')
    plt.show()
````

````python
# ejemplo del sitio
lotka_volterra(a = 0.7, b = 0.5, c = 0.3, e = 0.2, dt = 0.001, t0 = 0, max_time = 100, x0 = 1.0, y0 = 0.5)
````

## A)	Si aumenta considerablemente el aumento natural de las presas
Básicamente, el tiempo en el que se reproducen las presas es más rápido en comparación al caso original, lo que conlleva a que exista más alimento y hace que se vea reflejado en que los depredadores, pues estos pueden alimentarse más y crecer de forma proporcional a la población de las presas.

````python
lotka_volterra(a=1.1, b=0.5, c=0.3, e=0.2, dt=0.001, t0 = 0, max_time=100, x0=1.0, y0=0.5)
````

## B)	Si hay un cambio considerablemente en la destrucción por depredadores
Ahora con la gran habilidad que tienen los depredadores para cazar presas, significa que los depredadores son mucho más que la población de las presas, la población de los depredadores incrementará ya que su población tiene mas probabilidades de encontrar alimento y, por otro lado, dejará con muy pocas posibilidades de que se reproduzcan las presas, por ende, estará en niveles muy bajos de población.

````python
lotka_volterra(a=0.7, b=0.1, c=0.3, e=0.2, dt=0.001, t0 = 0, max_time=100, x0=1.0, y0=0.5)
````

## C)	Si aumenta considerablemente la muerte en ausencia de presas
En otras palabras, quiere decir mueren a un ritmo más rápido cuando dejan de haber presas, y técnicamente significa que las presas en algún punto dejaran reproducirse, lo que dará paso a  que los depredadores se extingan de forma rápida, dando paso a que las presas al no existir peligro puedan crecer rápidamente.


````python
lotka_volterra(a=0.7, b=0.5, c=0.9, e=0.2, dt=0.001, t0 = 0, max_time=100, x0=1.0, y0=0.5)
````

## D)	Si hay un aumento considerablemente alto por alimentación disponible.
Esto significa que los depredadores son más rápidos al poblar con poco alimento, por tanto, y por haber más depredadores, no crecerá tanto la población de las presas.

````python
lotka_volterra(a=0.7, b=0.5, c=0.3, e=1.2, dt=0.001, t0 = 0, max_time=100, x0=1.0, y0=0.5)
````

Para este caso en especial tuve que realizar un análisis matemático, y todo partiendo desde el concepto del punto de equilibrio, donde se igualan las ecuaciones a 0, con la finalidad de encontrar valores que los parámetros puedan obtener y las poblaciones se comporten como constantes.
