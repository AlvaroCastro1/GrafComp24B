### Se grafican puntos dentro del plano
con ayuda de `scratter` se colocan puntos en el plano

```python
x1=0
x2=80

y1=0
y2=80

#Definir ejes
plt.axis([x1,x2,y1,y2])

#Habilitar el gird
plt.axis('on')
plt.grid(True)
plt.title("Ejemplo")

dx=5
dy=5

#Grafica puntos
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):

        plt.scatter(x,y, s=1.5, color ='lightgray')
```

# Vector

# Actividad 1
A partir de los siguientes gráfica los vectores mostrando 3 posiciones posibles por cada uno
 *B* = (5,0)
 *C* = (0,-4)
 *D* = (-3,-5)

```python
# limites
y1 = -10
y2 = 10

x1 = -10
x2 = 10

# Definir ejes
plt.axis([x1, x2, y1, y2])

# para verlo de mejor forma, añadimos dos lineas que representen los ejes X y Y
plt.axhline(0, color='black',linewidth=1)  # Eje X
plt.axvline(0, color='black',linewidth=1)  # Eje Y

plt.axis('on')
plt.grid(True)
plt.title("Vectores 1ra Act")

dx = dy = 1

# Graficar puntos de la cuadrícula
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

# inicio en 
plt.arrow(-2, 1, 5, 0, head_length=0.5, head_width=0.3, color="b")
plt.arrow(-2, 2, 5, 0, head_length=0.5, head_width=0.3, color="b")
plt.arrow(-5, -2, 5, 0, head_length=0.5, head_width=0.3, color="b")

plt.arrow(0, 1, -4, 0, head_length=0.5, head_width=0.3, color="c")
plt.arrow(0, -3, -4, 0, head_length=0.5, head_width=0.3, color="c")
plt.arrow(6, 2.5, -4, 0, head_length=0.5, head_width=0.3, color="c")

plt.arrow(3, 5, -3, -5, head_length=0.5, head_width=0.3, color="k")
plt.arrow(3, 0, -3, -5, head_length=0.5, head_width=0.3, color="k")
plt.arrow(0, 0, -3, -5, head_length=0.5, head_width=0.3, color="k")
```

# Actividad 2
a partir de los siguientes gráfica los vectores mostrando 2 posibles opciones por cada uno 
M(2,4)
N(-3,3)
P(1,-4)

- MN
- PN

```python
# limites
y1 = -10
y2 = 10

x1 = -10
x2 = 10

# Definir ejes
plt.axis([x1, x2, y1, y2])

# para verlo de mejor forma, añadimos dos lineas que representen los ejes X y Y
plt.axhline(0, color='black',linewidth=1)  # Eje X
plt.axvline(0, color='black',linewidth=1)  # Eje Y

plt.axis('on')
plt.grid(True)
plt.title("Vectores Actividad 2")

dx = dy = 1

# Graficar puntos de la cuadrícula
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

# Graficar el vector desde (2, 4) hasta (-3, 3)
plt.arrow(2, 4, -5, -1, head_length=0.5, head_width=0.3, color="r")
plt.arrow(4, -2, -5, -1, head_length=0.5, head_width=0.3, color="r")

plt.arrow(1, -4, -4, 7, head_length=0.5, head_width=0.3, color="b")
plt.arrow(5, -5, -4, 7, head_length=0.5, head_width=0.3, color="b")
```

La importancia de esta funcion radica en que la funcion toma como `parametros` el **punto inicial** del `vector` y el **desplazamiento** que tendra en cada uno de los ejes
