 El **núcleo** de una matriz son todos los vectores que la matriz envía al punto 0.

```python
import numpy as np
from scipy.linalg import null_space

# Definir la matriz
A = np.array([[1, 1, 0],
              [1, 1, 0]])

# Calcular el núcleo de la matriz
nucleo = null_space(A)

print("Núcleo de la matriz:")
print(nucleo)
```

```python
A @ nucleo
```

```python
A = np.array([[1, 1, 2],
              [2, 2, 4],
              [2, 3, 5]])

# Calcular el núcleo de la matriz
nucleo = null_space(A)

print("Núcleo de la matriz:")
print(nucleo)
```
