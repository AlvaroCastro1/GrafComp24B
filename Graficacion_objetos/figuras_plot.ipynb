### bibliotecas necesarias para graficar con matplotlib

```python
import matplotlib.pyplot as plt
```

## creamos un nuevo plano con un punto y lo mostramos

```python
plt.plot([1], [1])
plt.show()
```

### el metodo **plot** recibe dos listas con las coordenadas para X & Y

```python
# lista con coord X & Y
plt.plot([1,1.5,2], [1,2,1])

plt.show()
```

```python
# figura primera lista con cordenadas X y despues Y
plt.plot(
    [1, 1.5, 2, 1],
    [1,   2, 1, 1]
    )
plt.show()
```

```python
#hacer rectangulo
plt.plot(
    [1, 1, 2, 2, 1],
    [1, 2, 2, 1, 1]
    )
plt.show()
```

```python
# rectangulo
plt.plot(
    [1, 1, 2, 2, 1],
    [1, 2, 2, 1, 1]
    )
# triangulo
plt.plot(
    [2, 3, 4, 2],
    [2, 3, 2, 2]
    )
plt.show()
```

## cambiar la escala

```python

escala = 3
#figura original
plt.plot([1,1,2,2,1],[1,2,2,1,1])
# aumentar escala
plt.plot([1*escala, 2*escala, 2*escala, 1*escala, 1*escala],[1*escala, 1*escala, 2*escala, 2*escala, 1*escala])
plt.show()  
```

## hacer rotacion

```python
import numpy as np
# puntos para un cuadrado
x = [1, 2, 2, 1, 1]
y = [1, 1, 2, 2, 1]
# dibujar
plt.plot(x,y)

# convertir en un angulo de 45°
tetha = np.radians(45)

# funciones trigonometricas np.sin y np.cos
rotacionX = np.cos(tetha) * np.array(x) - np.sin(tetha) * np.array(y)
rotacionY = np.sin(tetha) * np.array(x) + np.cos(tetha) * np.array(y)

#mostrar figura rotada
plt.plot(rotacionX, rotacionY)
plt.show()
```

## hacer traslacion

```python
# lista de punto
x = [1, 2, 2, 1, 1]
y = [1, 1, 2, 2, 1]

plt.plot(x, y)

# se le añade un valor para moverla en este caso 5
plt.plot([xi + 5 for xi in x],[yi + 5 for yi in y])
plt.show()
```
